# Sampling Distribution of the Mean

Here it is!  My first substantive post on AB testing in data science.  And I couldn't think of a better place to start than with an introduction to sampling distributions.  Let's work through this with the example of the *sampling distribution of the mean*. 

Let's say that we're working with a team that wants to roll out a new feature for an app.  We think users will like the new feature and therefore that app engagment will increase, but engineers are worried that the increased latency will turn people off, potentially causing a decrease in app engagement.  Because of this risk, we can't roll out the feature to all users.  So instead we decide to roll it out to a small, random sample of the user base.  We know that from this small sample we can *infer* the average engagment time of the entire population in the user base.  You probably even know the formulas to calculate a standard error and a confidence interval for this estimation.  But how and why does this work?  The concept of a sampling distribution is the key.  

First, some setup:



````r
set.seed(145)
options(tidyverse.quiet=TRUE)
suppressWarnings(library(glue))
suppressWarnings(library(tidyverse))
theme_set(theme_minimal())
````

## OK, let's start with our sample

Ok, so now we've drawn a sample, rolled out the new feature to this small group of users, and we've observed the engagement times of each user.   Now, let's assume that the true population mean (that is, the true average engagement time of **all users** if we had rolled out to all users) is 100 minutes.  We can also assume a true standard deviation, we'll set this to 10.  This gives us the population distribution from which we have drawn our sample.  From this distribution we will pull a sample of 100 users.  For this example we will assume that user engagement times are normally distributed.  Let's have a look at the engagement times we observed.  

````r
population_mean = 100
population_sd = 10
n_users = 100

x_sample = rnorm(n_users, population_mean, population_sd)

df = tibble(
    x=x_sample,
    id=1:n_users
)

df %>% 
  summarise(x_mean=mean(x_sample),
           x_sd=sd(x_sample))
````

In our sample we have observed a mean engagement time of 100.98 minutes, with a standard deviation of 10.13 minutes.  We can plot this data using a histogram with an estimaged density plot.  The data resembles somewhat the normal distribution from which the sample was drawn, but given that there are only 100 observations, is far from perfectly normal.  

````r
df %>%
  ggplot(aes(x=x)) +
  geom_histogram(aes(y=..density..), fill='steelblue', alpha=0.7, bins=30) + 
  geom_density(color='darkorange', size=1)
````

### Standard errors and confidence intervals

Now you have your sample.  Yay!  Now what?  How can you use this information to infer the engagement for the __entire population__ of users?  You all probably know what to do from here, the estimated mean in the population is just the mean of your sample, and the standard error of the estimate is calculated as the standard deviation of the sample divided by the square root of n:  

<h2><center> $\hat{\mu} = m$ </center></h2>
    
<h2><center> And </center></h2>
    
<h2><center> $se_{\hat{\mu}} = \frac{s}{\sqrt{n}}$ </center></h2>

From there we can calculate the confidence interval of our choice.  Here is the estimated engagement time and 95% confidence interval calculated from our sample:

````r
estimated_pop_mean = mean(x_sample)
standard_error_of_estimate = sd(x_sample) / sqrt(n_users)
lower_bound_95 = estimated_pop_mean - 1.96 * standard_error_of_estimate
upper_bound_95 = estimated_pop_mean + 1.96 * standard_error_of_estimate

tibble(
    experiment_id=as.factor(1),
    est_mean=estimated_pop_mean,
    low_bound=lower_bound_95,
    upp_bound=upper_bound_95
) %>%
  ggplot(aes(x=experiment_id, y=est_mean, ymin=low_bound, ymax=upp_bound)) +
  geom_pointrange(color='steelblue') + 
  labs(x='Sample ID', y='Estimated mean + 95% CI', title='Experiment Result') +
  annotate(x=1.12, y=estimated_pop_mean, geom='text', label=glue('Point estimate \n ({round(estimated_pop_mean, 2)})')) +
  annotate(x=1.07, y=lower_bound_95 + 0.05, geom='text', label=glue('LB of 95% \n ({round(lower_bound_95, 2)})')) +
  annotate(x=1.07, y=upper_bound_95 - 0.05, geom='text', label=glue('UB of 95% \n ({round(upper_bound_95, 2)})')) +
  coord_flip()
````

### But, how does this work?  What is a standard error?  And why do we calculate the confidence interval using it?  

Let's pretend that we didn't know how to calculate the standard error and confidence interval.  We pulled a single sample of 100 users, we looked at the distribution of data, and that was it.  Not knowing what to do with our data, we decide to re-run the experiment, and roll-out the new feature to a new sample of 100 users.  This time, we get a slightly different result to the first.  Since we got a slightly different result, we repeated the experiment _again_ and again and again.  The observed mean value of each of our samples is close to, but never exactly the same as, the mean engagement in the prior sample.  And all measurements are somewhat close to 100, give or take. 

````r
repeated_samples = replicate(50, rnorm(n_users, population_mean, population_sd))
repeated_samples %>% colMeans() %>% head()
````

Using ```replicate``` gives us a matrix of shape 100x50, 100 observations drawn 50 times.  Which means that we can get the sample means by taking the mean of each column.  We can also plot these mean values to see how they are distributed.  

````r
repeated_samples %>% 
  colMeans() %>% 
  as_tibble() %>%  
  mutate(sample_id = row_number())  %>%
  ggplot(aes(x=as.factor(sample_id), y=value)) + 
  geom_hline(yintercept=repeated_samples %>% colMeans() %>% mean(), linetype='dashed') +
  geom_point(color='steelblue', size=2) + 
  labs(y='Observed mean of sample', x='Sample ID') +
  coord_flip()
````

With our 50 samples in which we've observed the mean value for each, we decide to estimate the mean of our means.  And we can calculate the standard deviation of our means.  Let's see what we get.

````r
mean_of_sample_means = repeated_samples %>% colMeans() %>% mean()
sd_of_sample_means = repeated_samples %>% colMeans() %>% sd()

tibble(mean_of_sample_means, sd_of_sample_means)
````

You'll notice that these values are not too far off of the values we estimated from our first sample, the example in which we calculated the standard error parametrically:

````r
tibble(estimated_pop_mean, standard_error_of_estimate)
````

What we've just done by repeating our experiment 50 times is the start of a sampling distribution!  A sampling distribution is precisely this, it is the distribution of observed values of a statistic of interest if the experiment were to be repeated infinitely.  In our example we've collected only 50 samples, so we cannot really claim to have approximated a sampling distribution, but this is the general idea.  If we were to repeat our experiment an infinite number of times, and calculate the observed mean in each sample, these estimated means form a sampling distribution.  

Let's see what happens if we were to repeat our sampling 100,000 times instead of only 50:

````r
many_repeated_samples = replicate(1e5, rnorm(n_users, population_mean, population_sd))

tibble(
    estimated_mean=many_repeated_samples %>% colMeans() %>% mean(),
    sd_of_estimate=many_repeated_samples %>% colMeans() %>% sd()
)
````

Now we're getting very close to our true value (remember, we set those at the beginning of this example, the true mean is 100).  The more samples we draw, the closer the mean of the samples will be to the true mean.  

So, the sampling distribution of the mean is the theoretical distribution we would get if we repeatedly drew samples of a given size from the population and observed the mean value of our variable of interest in each sample.  The mean of the sampling distribution is the population mean in expectation.  And when we calculate a standard error from a single sample, what we are estimating is the standard deviation of the sampling distribution.  

# Central limit theorem

But here's the really cool thing about sampling distributions... it doesn't matter the shape of the underlying distribution from which the sample was drawn, the sampling distribution will always be normal!  This is hugely significant because it gives practical application of the concept of sampling distributions to data that is not normally distributed (because really, how many of DS actually work with normally distributed data???).  

There are a few exceptions to this rule... The underlying distribution must have a finite variance.  This means that the CLT does not apply if the underlying distribution is Cauchy for example.  Also, each of the samples in the distribution must be *independent*. 


Let's take a look at 3 different distributions to demonstrate how this works.  First, the normal distriburtion.  Here is histogram of 10,000 draws from a normal dristribution with mean 0 and sd 1.  It forms, roughly, the familiar bell curve.  This is analagous to pulling a sample of 10,000 people from your population of users (assuming that whatever behaviour they exhibit forms a normal distribution -- this is probably quite rare in reality).  In any case, in this example, the distribution of values of this sample forms, roughly, the familiar bell curve.  

````r
dists = tibble(
    poisson=rpois(1e4, 2),
    uniform=runif(1e4, 0, 1),
    normal=rnorm(1e4, 0, 1)
)
````

````r
dists %>%
  gather() %>%
  ggplot(aes(x=value)) + 
  geom_histogram(fill='steelblue', alpha=0.7, bins=30) + 
  facet_wrap(~key, scales='free')
````

````r
poisson_sampling_dist = replicate(1e5, rpois(100, 1))
uniform_sampling_dist = replicate(1e5, runif(100, -1, 3))
normal_sampling_dist = replicate(1e5, rnorm(100, 1, 1))

tibble(
    poisson=poisson_sampling_dist %>% colMeans(),
    uniform=uniform_sampling_dist %>% colMeans(),
    normal=normal_sampling_dist %>% colMeans()
) %>%
  gather() %>%
  ggplot(aes(x=value)) + 
  geom_histogram(fill='steelblue', bins=30, alpha=0.7) +
  facet_wrap(~key, scales='free')
````

Amazing, right?  The sampling distribution is approximately normal (approximately because we only took 100,000 samples, not an infinte number, from each) in all three cases.  Mind blown.

# So, who cares?


You might think, who cares really?  In practice nobody ever takes tens of thousands of samples to approximate a sampling distribution, so this is all simply theoretical.  And you would be right, sampling distributions are purely theoretical.  BUT, the importance of this cannot be understated.  Given a theoretical understanding of how the sampling distribution of the population might look, **we can infer the likelihood that our sample will be similar to other samples**.  And if we can infer that, then we can infer the likelihood that our sample will be similar to the population that we are interested in.  In other words, the sampling distribution lets us move from descriptive to inferential statistics -- the sampling distribution is the foundation of inference!  This is, of course, what we aim to do in any experiment or quantitative study, infer from a sample to a population.  Further, the finding that the sampling distribution will be approximately normal regardless of the underlying distribution (the CLT) gives much wider practical application.  Because, really, you're not likely to find many bell curves in the wild...  And while I focused on the sampling distribution of the mean in this post, there exists a sampling distribution for _any test statistic_!  Which opens up for questions like, if the correlation between two variables in a sample is $r$, within what range is the correlation in the population likely to be?  But that will have to wait for my next post.  

Happy sampling!
