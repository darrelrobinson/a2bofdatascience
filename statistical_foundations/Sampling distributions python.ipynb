# Sampling distribution of the mean

Python translation of R code in blog post, 'What is a sampling distribution?'

```python
import pandas as pd
import numpy as np
np.random.seed(0)

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
population_mean = 100
population_sd = 10
n_users = 100

x_sample = np.random.normal(population_mean, population_sd, n_users)
print(f'Sample mean = {x_sample.mean()} \t Sample sd = {x_sample.std()}')
```

```python
sns.histplot(x_sample, bins=30, kde=True);
```

```python
estimated_pop_mean = x_sample.mean()
standard_error_of_estimate = x_sample.std() / np.sqrt(n_users)
lower_bound_95 = estimated_pop_mean - 1.96 * standard_error_of_estimate
upper_bound_95 = estimated_pop_mean + 1.96 * standard_error_of_estimate

plt.plot(estimated_pop_mean, 1, 'o')
plt.hlines(y=1, xmin=lower_bound_95, xmax=upper_bound_95)
plt.annotate(f'lower bound \n {round(lower_bound_95, 2)}', (lower_bound_95 + 0.05, 1.005))
plt.annotate(f'upper bound \n {round(upper_bound_95, 2)}', (upper_bound_95 - 0.5, 1.005))
plt.annotate(f'point estimate \n {round(estimated_pop_mean, 2)}', (100.3, 1.015));
```

```python
repeated_samples = []
for i in range(50):
    repeated_samples.append(np.random.normal(population_mean, population_sd, n_users))
    
[i.mean() for i in repeated_samples][:5]
```

```python
sample_means = [i.mean() for i in repeated_samples]
pd.DataFrame({'sample means': sample_means}).reset_index().plot.scatter(y='index', x='sample means')
plt.vlines(x=np.array(sample_means).mean(), ymin=0, ymax=50, color='gray', linestyles='dashed')
```

```python
mean_of_sample_means = np.array(sample_means).mean()
sd_of_sample_means = np.array(sample_means).std()

print(f'mean of sample means: {round(mean_of_sample_means, 2)} \t sd of sample means: {round(sd_of_sample_means, 2)}')
```

```python
print(f'estimated population mean: {round(estimated_pop_mean, 2)} \t standard error of estimate: {round(standard_error_of_estimate, 2)}')
```

```python
many_repeated_samples = []

for i in range(100000):
    many_repeated_samples.append(np.random.normal(population_mean, population_sd, n_users))

mean_of_many_means = np.array([i.mean() for i in many_repeated_samples]).mean()
sd_of_many_means = np.array([i.mean() for i in many_repeated_samples]).std()

print(f'mean of many means: {round(mean_of_many_means, 4)} \t sd of means {round(sd_of_many_means, 4)}')
```

# Central limit theorem

```python
pois = np.random.poisson(2, 10000)
unif = np.random.uniform(0, 1, 10000)
norm = np.random.normal(0, 1, 10000)

df = pd.DataFrame({'poisson':pois, 'uniform':unif, 'normal':norm}).melt()

p = sns.FacetGrid(df, col='variable', sharex=False, sharey=False)
p.map_dataframe(sns.histplot, 'value');
```

```python
poisson_samples = []
uniform_samples = []
normal_samples = []
for i in range(100000):
    poisson_samples.append(np.random.poisson(1, 100))
    uniform_samples.append(np.random.uniform(-1, 3, 100))
    normal_samples.append(np.random.normal(1, 1, 100))


sampling_dists = pd.DataFrame({
    'poisson': [i.mean() for i in poisson_samples],
    'uniform': [i.mean() for i in uniform_samples],
    'normal': [i.mean() for i in normal_samples]
}).melt()
```

```python
p = sns.FacetGrid(sampling_dists, col='variable', sharex=False, sharey=False)
p.map_dataframe(sns.histplot, 'value');
```
